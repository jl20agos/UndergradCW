```python
from astropy.io import ascii
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
file = 'starmodl.dat'
star = ascii.read(file,comment='#',delimiter=' ')
star
```

```python
plt.figure(figsize=(8,6))
plt.plot(star['r'],star['T'],label='Temperature')
plt.xlabel('Radius (cm)',fontsize=16)
plt.ylabel('Temperature (K)',fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
ax = plt.gca()
ax.set_xscale('log')
#ax.st
```

```python
plt.figure(figsize=(8,6))
plt.plot(star['r'],star['P'],label='Temperature')
plt.xlabel('Radius (cm)',fontsize=16)
plt.ylabel('Pressure (atm)',fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
ax = plt.gca()
ax.set_xscale('log')
#ax.st
```

```python
plt.figure(figsize=(8,6))
plt.plot(star['r'],star['rho'],label='Density')
plt.xlabel('Radius (cm)',fontsize=16)
plt.ylabel('Density (g/$cm^3$)',fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
ax = plt.gca()
ax.set_xscale('log')
#ax.st
```

```python
plt.figure(figsize=(8,6))
plt.plot(star['r'],star['L_r'],label='Luminosity')
plt.xlabel('Radius (cm)',fontsize=16)
plt.ylabel('Luminosity (J/s)',fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
ax = plt.gca()
ax.set_xscale('log')
#ax.st
```
